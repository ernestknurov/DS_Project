```python
import numpy as np 
import pandas as pd
import matplotlib as mtl
```

## ETL check

```python
def extract(paths: dict) -> tuple:

    """Extract data from a given paths"""

    item_cat = pd.read_pickle(paths['item_cat'])
    items = pd.read_pickle(paths['items'])
    sales_train = pd.read_pickle(paths["sales_train"])
    shops = pd.read_pickle(paths["shops"])
    test = pd.read_pickle(paths["test"])
    return item_cat, shops, items, sales_train, test
```

```python
paths_to_cleaned_data = {
    'item_cat': "cleaned_data/item_cat.pickle",
    "items": "cleaned_data/items.pickle",
    "sales_train": "cleaned_data/sales_train.pickle",
    "shops": "cleaned_data/shops.pickle",
    "test": "cleaned_data/test.pickle"
}
```

```python
%%time
item_cat, shops, items, sales_train, test = extract(paths_to_cleaned_data)
```
